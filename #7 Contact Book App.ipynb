# Python Contact Book Application

This project demonstrates a menu-driven contact book built in Python.

The application allows the user to:
- add contacts
- update contacts
- delete contacts
- view all contacts
- search contacts by name or number

## Concepts practiced
- dictionaries
- functions
- loops
- conditionals
- CRUD operations
- duplicate validation

## Project Overview

This project simulates a simple contact management system.

Each contact is stored in a dictionary where:
- the key is the contact name
- the value is the phone number

The system also includes validation to prevent:
- duplicate contact names
- duplicate phone numbers

## Step 1: Create the Contact Data Dictionary

The application begins with a dictionary that stores sample contacts and their phone numbers.

```python
# Contact data

contacts = {
    "alice": "4155551021",
    "bob": "4155551034",
    "charlie": "4155551045",
    "david": "4155551056",
    "emma": "4155551067",
    "frank": "4155551078",
    "grace": "4155551089",
    "henry": "4155551090",
    "isabella": "4155551101",
    "jack": "4155551112"
}
```

## Step 2: Add Contact Function

This function:
- accepts a new contact name and number
- checks whether the name already exists
- checks whether the number already exists
- adds the contact if both are unique

```python
def add_contact(contacts):
    input_name = input("Name: ").strip().lower()
    input_number = input("Number: ").strip()

    if input_name in contacts:
        print("Contact already exists.")
        print(f"{input_name}: {contacts[input_name]}")
        return

    if input_number in contacts.values():
        print("This number already exists under:")

        for name, number in contacts.items():
            if number == input_number:
                print(f"{name}: {number}")
        return

    contacts[input_name] = input_number
    print("Contact added!")
```

## Step 3: Update Contact Function

This function updates the phone number of an existing contact.

```python
def update_contact(contacts):
    input_name = input("Which contact do you want to update? ").strip().lower()
    input_number = input("Please enter the number: ").strip()

    if input_name in contacts:
        contacts[input_name] = input_number
        print("Contact updated!")
        print(f"{input_name}: {contacts[input_name]}")
    else:
        print("Contact not found.")
```

## Step 4: Delete Contact Function

This function allows the user to delete a contact by:
- name
- phone number

```python
def del_contact(contacts):
    user_del = input("Enter name or number to delete: ").strip().lower()

    # Case 1: user entered a name
    if user_del in contacts:
        del contacts[user_del]
        print("Contact deleted.")

    # Case 2: user entered a number
    elif user_del in contacts.values():
        for name, number in contacts.items():
            if number == user_del:
                del contacts[name]
                print("Contact deleted.")
                break

    else:
        print("Contact not found.")
```

## Step 5: View Contacts Function

This function displays all saved contacts.

```python
def view_contact(contacts):
    if not contacts:
        print("Contacts list is empty.")
        return

    print("\nCurrent Contacts:")

    for name, number in contacts.items():
        print(f"{name}: {number}")
```

## Step 6: Search Contact Function

This function allows the user to search for a contact using:
- the contact name
- the phone number

```python
def search_contact(contacts):
    user_search = input("What name/number do you want to search? ").strip().lower()

    # Case 1: search by name
    if user_search in contacts:
        print(f"{user_search}: {contacts[user_search]}")

    # Case 2: search by number
    elif user_search in contacts.values():
        for name, number in contacts.items():
            if number == user_search:
                print(f"{name}: {number}")
                break

    else:
        print("Contact not found.")
```

## Step 7: Main Application Loop

The menu system connects all the functions and allows the user to manage contacts until they choose to exit.

```python
while True:
    print("""
What do you want to do with your contacts?
1. Add contact
2. Update contact
3. Delete contact
4. View contacts
5. Search contact
6. Exit
""")

    user_input = input("Choose your option: ")

    if user_input == "1":
        add_contact(contacts)

    elif user_input == "2":
        update_contact(contacts)

    elif user_input == "3":
        del_contact(contacts)

    elif user_input == "4":
        view_contact(contacts)

    elif user_input == "5":
        search_contact(contacts)

    elif user_input == "6":
        print("Goodbye!")
        break

    else:
        print("Invalid option. Please try again.")
```
