# Python Mini Calculator

This project demonstrates a simple calculator application built step-by-step.

The calculator supports:
- Addition
- Subtraction
- Multiplication
- Division

It evolves from a single calculation script to a loop-based interactive application with input validation.

## Project Overview

This project was developed in three stages:

1. Basic calculator (single calculation)
2. Loop-based calculator (continuous usage)
3. Enhanced calculator with:
   - division-by-zero handling
   - user-controlled continuation

The goal is to practice control flow, loops, and user interaction.

## Step 1: Basic Calculator (Single Calculation)

In this version, the user performs one calculation based on the selected operator.

```python
# Step 1: Basic calculator

num_1 = float(input("Enter the first number: "))
operator = input("Choose an operator (+, -, *, /): ")
num_2 = float(input("Enter the second number: "))

if operator == "+":
    print(num_1 + num_2)

elif operator == "-":
    print(num_1 - num_2)

elif operator == "*":
    print(num_1 * num_2)

elif operator == "/":
    print(num_1 / num_2)

else:
    print("Invalid operator")
```

### Logic Used

- Took numerical input from the user
- Used conditional statements to perform operations
- Displayed result based on selected operator

## Step 2: Add Loop for Continuous Use

This version allows the user to perform multiple calculations using a loop.

```python
# Step 2: Loop-based calculator

while True:
    print("Welcome to the mini-calculator!")

    num_1 = float(input("Enter the first number: "))
    operator = input("Choose an operator (+, -, *, /): ")
    num_2 = float(input("Enter the second number: "))

    if operator == "+":
        print(f"Your sum is: {num_1 + num_2}")

    elif operator == "-":
        print(f"Your difference is: {num_1 - num_2}")

    elif operator == "*":
        print(f"Your product is: {num_1 * num_2}")

    elif operator == "/":
        print(f"Your division is: {num_1 / num_2}")

    else:
        print("Invalid operator")
        break
```

### Logic Used

- Introduced a loop to allow repeated calculations
- Continued execution until an invalid operator is entered

## Step 3: Enhanced Calculator

This version improves the application by:
- preventing division by zero
- allowing the user to decide whether to continue

```python
# Step 3: Final calculator with validation

while True:
    print("Welcome to the mini-calculator!")

    num_1 = float(input("Enter the first number: "))
    operator = input("Choose an operator (+, -, *, /): ")
    num_2 = float(input("Enter the second number: "))

    if operator == "+":
        print(f"Your sum is: {num_1 + num_2}")

    elif operator == "-":
        print(f"Your difference is: {num_1 - num_2}")

    elif operator == "*":
        print(f"Your product is: {num_1 * num_2}")

    elif operator == "/":
        if num_2 == 0:   
            print("Cannot divide by 0")
        else:
            print(f"Your division is: {num_1 / num_2}")

    else:
        print("Invalid operator")
        break

    user_input = input("Do you want to calculate again? (y/n): ")

    if user_input != "y":
        print("Exiting")
        break
```

## Example Usage

- User enters: 5 + 3 → Output: 8  
- User enters: 10 / 0 → Output: Cannot divide by 0  
- User chooses 'n' → Application exits

## Key Learnings

- Implemented arithmetic operations using conditionals
- Used loops for continuous interaction
- Handled edge cases like division by zero
- Improved user experience with continuation logic
