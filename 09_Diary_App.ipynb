# Notes / Diary App

This notebook demonstrates a simple diary application built in Python.

The project is shown in two versions:

1. **V1** – A basic diary app where entries are stored temporarily during program execution  
2. **V2** – An improved version where entries are saved to a JSON file for persistence

## Concepts practiced
- functions
- lists
- loops
- conditionals
- file handling
- JSON storage

## Project Overview

This project simulates a diary or notes application where a user can:

- add a diary entry
- view saved entries
- exit the application

The first version stores entries only while the program is running.

The second version improves the application by saving entries to a JSON file, so that diary entries remain available even after restarting the program.

## Version 1: Basic Diary App

This first version builds the core structure of the application.

### What this version does

This version builds a simple diary application where:

- users can add entries
- users can view entries
- entries are stored temporarily in a list

### How it works

- Functions are created to:
  - display the menu
  - add entries
  - view entries
- A list is used to store diary entries
- A loop keeps the application running until the user exits

```python
def show_menu():
    print("\n--- Diary App ---")
    print("1. Add Entry")
    print("2. View Entries")
    print("3. Exit")


def add_entry(entries, text):
    entries.append(text)
    print("Entry added successfully.")


def view_entries(entries):
    if not entries:
        print("No diary entries yet.")
        return

    print("\n--- Your Diary Entries ---")
    for index, entry in enumerate(entries, start=1):
        print(f"{index}. {entry}")
```

```python
def main():
    entries = []

    while True:
        show_menu()
        choice = input("Choose an option: ")

        if choice == "1":
            text = input("Write your diary entry: ")
            add_entry(entries, text)

        elif choice == "2":
            view_entries(entries)

        elif choice == "3":
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")

main()
```

## Version 2: Saving Entries to JSON

The second version improves the diary app by adding file storage.

Instead of keeping entries only in memory, entries are now saved in a JSON file.

This makes the application more practical because the data remains available even after the program is closed and reopened.

### How it works

1. Functions are defined first:
   - `show_menu()` displays options
   - `load_entries()` reads data from the JSON file
   - `save_entries()` writes data to the file
   - `add_entry()` adds a new entry (with validation)
   - `view_entries()` displays all entries

2. Data is stored in a list:
   - entries are loaded from `diary.json`
   - new entries are added to this list

3. File handling:
   - data is saved using JSON format
   - if the file does not exist, an empty list is used

This version ensures that diary entries are stored permanently.

```python
import json

FILE_NAME = "diary.json"


def show_menu():
    print("\n--- Diary App ---")
    print("1. Add Entry")
    print("2. View Entries")
    print("3. Exit")


def load_entries():
    # Load entries from JSON file
    try:
        with open(FILE_NAME, "r") as file:
            return json.load(file)
    except FileNotFoundError:
        # If file doesn't exist, return empty list
        return []


def save_entries(entries):
    # Save entries to JSON file
    with open(FILE_NAME, "w") as file:
        json.dump(entries, file, indent=4)


def add_entry(entries, text):
    # Prevent empty entries
    if text.strip() == "":
        print("Entry cannot be empty.")
        return False

    entries.append(text)
    return True


def view_entries(entries):
    # Display all entries
    if not entries:
        print("No diary entries yet.")
        return

    print("\n--- Your Diary Entries ---")
    for index, entry in enumerate(entries, start=1):
        print(f"{index}. {entry}")
```

```python
def main():
    entries = load_entries()

    while True:
        show_menu()
        choice = input("Choose an option: ")

        if choice == "1":
            text = input("Write your diary entry: ")

            was_added = add_entry(entries, text)
            if was_added:
                save_entries(entries)
                print("Entry added successfully.")

        elif choice == "2":
            view_entries(entries)

        elif choice == "3":
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")


main()
```
