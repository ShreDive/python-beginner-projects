The notebook covers:

## Concept Understanding

A Caesar Cipher shifts each letter in a message by a fixed number.

Example:

- Original: `abc`
- Shift by 2
- Encrypted: `cde`

This notebook builds that logic step by step, starting from the most basic version and then improving it.

## ASCII Basics

Before building the cipher, I first understood how characters and numbers are connected in Python.

Examples:
- `a` → 97
- `b` → 98
- `c` → 99
- ...
- `z` → 122

This is why `ord()` and `chr()` are important in this project.

## Caesar Cipher Structure

The clean structure of the Caesar Cipher is:

- Input
- Setup
- Process
- Output

```python
message = input("Enter your message: \n")
shift = int(input("by what number would you like to shift your message? \n"))
```

```python
# store the result as:

encrypted_message = ""
```

## Step 1: Simple Encryption

This first version performs a direct shift on every character in the message.

At this stage, spaces and symbols are not handled separately.
The goal here was to first understand the core shifting logic.

```python
# Step 1: loop through the letters:
for letter in message:
    new_letter = chr(ord(letter) + shift)
    encrypted_message += new_letter
    
# print the output
print("Your encrypted message :\n ", encrypted_message)
```

## Step 2: Handle Spaces, Punctuation, and Numbers

In this version, I improved the logic so that only alphabet characters are shifted.

If a character is not a letter, it is kept unchanged.
This allows spaces, punctuation, and numbers to remain as they are.

```python
# Step 2
encrypted_message = ""

for letter in message:
    if letter.isalpha():
        new_letter = chr(ord(letter) + shift)
    else:
        new_letter = letter

    encrypted_message += new_letter

print("Your encrypted message:\n", encrypted_message)
```

## Step 3: Use Modulo to Stay Within Alphabet Letters

The previous version could still produce symbols like `{`, `}`, or other non-letter characters when shifting beyond `z` or `Z`.

To fix that, I used modulo logic.

This ensures that:
- `z` wraps back to `a`
- `Z` wraps back to `A`

This is the final correct Caesar Cipher logic for letters.

```python
# Step 3 : modulo
encrypted_message = ""

for letter in message:
    if letter.islower():
        new_letter = chr((ord(letter) - ord('a') + shift) % 26 + ord('a'))
    elif letter.isupper():
        new_letter = chr((ord(letter) - ord('A') + shift) % 26 + ord('A'))
    else:
        new_letter = letter

    encrypted_message += new_letter

print("Your encrypted message:\n", encrypted_message)
```

## Final Version for User Input

After building the logic step by step, I converted it into reusable functions so that the user can choose between:

- encryption
- decryption

```python
# Final Caesar Cipher (clean version)

while True:
    print("""
Choose an option:
1. Encrypt
2. Decrypt
3. Exit
""")

    choice = input("Enter your choice: ")

    if choice == "3":
        print("Exiting...")
        break

    message = input("Enter your message: ")
    shift = int(input("Enter shift value: "))

    result = ""

    # Decryption = negative shift
    if choice == "2":
        shift = -shift

    for letter in message:

        if letter.islower():
            new_letter = chr((ord(letter) - ord('a') + shift) % 26 + ord('a'))

        elif letter.isupper():
            new_letter = chr((ord(letter) - ord('A') + shift) % 26 + ord('A'))

        else:
            new_letter = letter

        result += new_letter

    if choice == "1":
        print("Encrypted message:", result)

    elif choice == "2":
        print("Decrypted message:", result)

    else:
        print("Invalid option")
```
