```python
# Basic Expense tracker V1

total_expense = 0

while True :
    user_expense = (input("Describe your expense: "))
    user_amount = float(input("Amount: "))

    total_expense = user_amount + total_expense
    print("Expense added!")
    print(f"Current total spent : {total_expense}")
```

```python
# Expense tracker V2 

total_expense = 0
expenses = [] 

while True :
    print("""
    1. Add expense
    2. View total spent
    3. View all expenses
    4. Exit
    """)
    
    user_input = int(input("Option: "))
    
    if user_input == 1:
        user_expense = (input("Describe your expense: "))
        user_amount = float(input("Amount: "))
        
        total_expense = user_amount + total_expense
        expenses.append((user_expense, user_amount))
            
        print("Expense added!")
        print(f"Current total spent : {total_expense}")
        
    elif user_input == 2:
        print(f"Your total spent is {total_expense}")
    
    elif user_input == 3: 
        print("Your expense: ")
        for item in expenses :
            print(f"- {item[0]} : {item[1]}")
    
    elif user_input == 4:
        print("Exiting")
        break
```
