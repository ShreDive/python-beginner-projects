# Python Inventory Management System

This project demonstrates a menu-driven inventory management application built in Python.

The application allows the user to:
- add items
- update item quantities
- remove items
- view current inventory

It also includes a synonym dictionary to handle:
- plural forms
- common misspellings
- typos

## Concepts practiced
- dictionaries
- functions
- loops
- conditionals
- string cleaning
- data normalization

## Project Overview

This project simulates a simple inventory management system for fruit items.

A synonym dictionary is used to map different user inputs to a standard item name.
For example:
- `apples` → `apple`
- `bannana` → `banana`
- `ornage` → `orange`

This makes the application more flexible and user-friendly.

## Step 1: Create the Data Dictionaries

The project begins with two dictionaries:

1. `synonyms`  
   Maps user-entered variations of item names to a canonical form

2. `inventory`  
   Stores the available quantity for each item

```python
# Data dictionaries

synonyms = {
    "apple": "apple",
    "apples": "apple",
    "appel": "apple",
    "aplpe": "apple",

    "banana": "banana",
    "bananas": "banana",
    "bnana": "banana",
    "bannana": "banana",
    "bananna": "banana",

    "mango": "mango",
    "mangoes": "mango",
    "mangos": "mango",
    "mengo": "mango",

    "orange": "orange",
    "oranges": "orange",
    "ornage": "orange",
    "orage": "orange",

    "grape": "grape",
    "grapes": "grape",
    "grap": "grape",
    "grpae": "grape"
}

inventory = {
    "apple": 10,
    "banana": 5,
    "mango": 12,
    "orange": 8,
    "grape": 15,
    "pineapple": 4,
    "watermelon": 6,
    "strawberry": 20,
    "blueberry": 14,
    "peach": 9,
    "pear": 7,
    "kiwi": 11
}
```

## Step 2: Add Item Function

This function:
- takes user input
- cleans it using `strip()` and `lower()`
- checks the synonym dictionary
- either updates an existing item or adds a new one

```python
def add_item(inventory, synonyms):
    raw = input("What item to add? ")
    clean = raw.strip().lower()
    canonical = synonyms.get(clean, clean)

    user_add_quantity = int(input("What quantity? "))

    if canonical in inventory:
        inventory[canonical] += user_add_quantity
        print("Quantity updated!")
        print(f"{canonical}: {inventory[canonical]}")
    else:
        inventory[canonical] = user_add_quantity
        print("Item added!")
        print(f"{canonical}: {inventory[canonical]}")
```

## Step 3: Update Item Function

This function replaces the quantity of an existing item.
If the item does not exist, it displays an error message.

```python
def update_item(inventory, synonyms):
    raw = input("Which item do you want to update? ")
    clean = raw.strip().lower()
    canonical = synonyms.get(clean, clean)

    user_update_quantity = int(input("Quantity: "))

    if canonical in inventory:
        inventory[canonical] = user_update_quantity
        print("Quantity updated!")
        print(f"{canonical}: {inventory[canonical]}")
    else:
        print("Item not found.")
```

## Step 4: Remove Item Function

This function removes an item from the inventory if it exists.

```python
def remove_item(inventory, synonyms):
    raw = input("Which item do you want to remove? ")
    clean = raw.strip().lower()
    canonical = synonyms.get(clean, clean)

    if canonical in inventory:
        del inventory[canonical]
        print(f"{canonical} removed from inventory.")
    else:
        print("Item not found.")
```

## Step 5: View Inventory Function

This function displays all items and quantities currently stored in the inventory.

```python
def view_inventory(inventory):
    if not inventory:
        print("Inventory is empty.")
        return

    print("\nCurrent Inventory:")

    for item, quantity in inventory.items():
        print(f"{item}: {quantity}")
```

## Step 6: Main Application Loop

The menu system connects all functions together and allows the user to interact with the inventory until they choose to exit.

```python
while True:
    print("""
1. Add item
2. Update item
3. Remove item
4. View inventory
5. Exit
""")

    user_input = input("Choose your option: ")

    if user_input == "1":
        add_item(inventory, synonyms)

    elif user_input == "2":
        update_item(inventory, synonyms)

    elif user_input == "3":
        remove_item(inventory, synonyms)

    elif user_input == "4":
        view_inventory(inventory)

    elif user_input == "5":
        print("Goodbye!")
        break

    else:
        print("Invalid option")
```
