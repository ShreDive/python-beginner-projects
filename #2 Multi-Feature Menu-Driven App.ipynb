# Menu-Driven Python Application

This project demonstrates a console-based menu-driven application built step-by-step.

Features included:
- Greeting a user
- Adding two numbers
- Checking if a string is a palindrome
- Exit functionality

This project focuses on:
- loops
- conditionals
- user input handling
- string manipulation
- incremental feature building

## Project Approach

The application was built incrementally in the following steps:

1. Created a basic menu structure (skeleton)
2. Added greeting functionality
3. Added number addition feature
4. Added palindrome checker

Each step improves the functionality of the application.

## Step 1: Application Skeleton

A basic menu structure using a `while` loop and conditional branching.

```python
run_app = True

while run_app:
    print("""
Welcome! Choose an option:
1. Greet user
2. Add two numbers
3. Check if a string is a palindrome
4. Exit
""")

    user_input = input("Enter your choice: ")

    if user_input == "1":
        pass
    elif user_input == "2":
        pass
    elif user_input == "3":
        pass
    elif user_input == "4":
        print("Exiting")
        break
    else:
        print("Invalid choice")
```

### Logic
- Used a `while` loop to continuously run the application
- Displayed menu options using a multi-line string
- Used `if-elif` statements to handle user choices

## Step 3: Add Number Addition Feature

```python
run_app = True

while run_app:
    print("""
Welcome! Choose an option:
1. Greet user
2. Add two numbers
3. Check if a string is a palindrome
4. Exit
""")

    user_input = input("Enter your choice: ")

    if user_input == "1":
        name = input("Enter your name: ")
        print(f"Hello {name}")
    elif user_input == "2":
        num_1 = int(input("Enter a number: "))
        num_2 = int(input("Enter a number: "))
        total = num_1 + num_2
        print(f"The sum is: {total}")
    elif user_input == "3":
        pass
    elif user_input == "4":
        print("Exiting")
        break
    else:
        print("Invalid choice")
```

## Step 4: Add Palindrome Checker

```python
run_app = True

while run_app:
    print("""
Welcome! Choose an option:
1. Greet user
2. Add two numbers
3. Check if a string is a palindrome
4. Exit
""")

    user_input = input("Enter your choice: ")

    if user_input == "1":
        name = input("Enter your name: ")
        print(f"Hello {name}")

    elif user_input == "2":
        num_1 = int(input("Enter a number: "))
        num_2 = int(input("Enter a number: "))
        total = num_1 + num_2
        print(f"The sum is: {total}")

    elif user_input == "3":
        word = input("Enter a word: ")
        reverse_word = ""

        for i in range(len(word) - 1, -1, -1):
            reverse_word += word[i]

        if reverse_word == word:
            print("Palindrome")
        else:
            print("Not Palindrome")

    elif user_input == "4":
        print("Exiting")
        break

    else:
        print("Invalid choice")
```
