# Python Expense Tracker

This project demonstrates a simple expense tracking application built step-by-step.

The application starts with a basic version that tracks total spending and evolves into a menu-driven system that stores and displays individual expenses.

## Concepts practiced
- variables and data types
- loops
- lists and tuples
- user input handling
- aggregation logic

## Project Overview

This project was developed in two stages:

1. Version 1: Tracks total expenses in a continuous loop
2. Version 2: A menu-driven application that allows:
   - adding expenses
   - viewing total spending
   - viewing all recorded expenses

The goal is to simulate a simple personal finance tracking system.

## Version 1: Basic Expense Tracker

In this version, the user continuously enters expenses, and the program updates the total spending.

There is no menu or exit option — the loop runs indefinitely.

```python
# Version 1: Basic expense tracker

total_expense = 0

while True:
    user_expense = input("Describe your expense: ")
    user_amount = float(input("Amount: "))

    total_expense += user_amount

    print("Expense added!")
    print(f"Current total spent: {total_expense}")
    break
```

### Logic Used

- Maintained a running total of expenses
- Took user input for expense description and amount
- Used a loop to continuously add expenses
- Updated total dynamically after each entry

## Version 2: Menu-Driven Expense Tracker

This version improves the application by adding a menu system.

Features:
- Add an expense
- View total spending
- View all expenses
- Exit the application

Expenses are stored in a list as (description, amount) pairs.

```python
# Version 2: Menu-driven expense tracker

total_expense = 0
expenses = []

while True:
    print("""
1. Add expense
2. View total spent
3. View all expenses
4. Exit
""")

    user_input = int(input("Option: "))

    if user_input == 1:
        user_expense = input("Describe your expense: ")
        user_amount = float(input("Amount: "))

        total_expense += user_amount
        expenses.append((user_expense, user_amount))

        print("Expense added!")
        print(f"Current total spent: {total_expense}")

    elif user_input == 2:
        print(f"Your total spent is {total_expense}")

    elif user_input == 3:
        print("Your expenses:")
        for item in expenses:
            print(f"- {item[0]}: {item[1]}")

    elif user_input == 4:
        print("Exiting")
        break

    else:
        print("Invalid option")
```

### Logic Used

- Used a list to store all expenses
- Stored each expense as a tuple (description, amount)
- Implemented a menu using a loop and conditionals
- Allowed users to view total and individual expenses
- Used `break` to exit the application

## Key Learnings

- Built a basic financial tracking system
- Learned how to store structured data using lists and tuples
- Practiced looping through data
- Implemented a menu-driven CLI application
- Managed state using variables (total_expense)
