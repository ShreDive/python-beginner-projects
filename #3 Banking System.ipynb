# Python Banking System

This notebook demonstrates a simple banking system built step by step.

The project begins with a basic single-operation banking script and then expands into a menu-driven banking application that supports multiple actions.

## Concepts practiced
- variables
- conditionals
- loops
- user input
- basic transaction logic

## Project Overview

This project was built in two stages:

1. A basic banking system where the user can either deposit or withdraw once
2. A menu-driven banking application where the user can repeatedly check balance, deposit, withdraw, or exit

The goal of this project is to practice control flow and user interaction in Python.

## Step 1: Basic Banking System

In this version, the user chooses whether to deposit or withdraw money.
The balance starts at 0, and only one transaction is performed.

```python
# STEP 1: Basic banking system

balance = 0
user_input = input("Do you want to deposit or withdraw? ")

if user_input == "deposit":
    amount = int(input("How much? "))
    balance = balance + amount

elif user_input == "withdraw":
    amount = int(input("How much? "))
    if amount <= balance:
        balance = balance - amount
    else:
        print("Insufficient balance")

else:
    print("Invalid. Please choose the correct option.")

print(f"Final balance: {balance}")
```

### Logic Used

- Created a starting balance of 0
- Asked the user whether they wanted to deposit or withdraw
- Updated the balance based on the selected action
- Prevented withdrawal when the balance was too low

## Step 2: Multi-Operation Banking App

This version improves the project by allowing the user to perform multiple banking actions in a loop.

Available options:
- Check balance
- Deposit money
- Withdraw money
- Exit the application

```python
# STEP 2: Multi-operation banking app

name = input("Name: ")
balance = 0

while True:
    print(f"""
Hello {name}. What would you like to do?
1. Check balance
2. Deposit
3. Withdraw
4. Exit
""")

    user_input = int(input("Option: "))

    if user_input == 1:
        print(f"Your current balance is {balance}")

    elif user_input == 2:
        amount = int(input("How much? "))
        balance = balance + amount
        print(f"Final balance: {balance}")

    elif user_input == 3:
        amount = int(input("How much? "))
        if amount <= balance:
            balance = balance - amount
            print(f"Final balance: {balance}")
        else:
            print("Insufficient balance")

    elif user_input == 4:
        print("Exiting")
        break

    else:
        print("Invalid option")
```

### Logic Used

- Used a `while True` loop to keep the application running
- Added a menu of banking actions
- Allowed repeated deposits and withdrawals
- Included a balance check option
- Used `break` to exit the application
