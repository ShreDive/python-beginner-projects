## Step 1: Fetch Weather Data from API

In this step, we connect to a weather API and request data for a given city.

The API returns raw JSON data, which contains detailed weather information.

### How it works

- A request is sent to the weather API using the city name
- Parameters like API key and city are passed in the request
- The API returns data in JSON format
- This data is stored as `raw_data` for further processing

```python
#1: Imports & constants
import requests

BASE_URL = "https://api.openweathermap.org/data/2.5/weather"
UNITS = "metric"
```

```python
#2: Function to fetch weather & add error handling
def get_weather(city, api_key):
    params = {
        "q": city,
        "appid": api_key,
        "units": UNITS
    }

    try:
        response = requests.get(BASE_URL, params=params, timeout=10)
        response.raise_for_status()
        return response.json()
    except requests.RequestException as error:
        print(f"Error fetching weather data: {error}")
        return None
```

## Step 2: Call the API Function

Here, the function is used to fetch weather data for a specific city.

### What this does

- Calls the API function
- Stores the returned JSON response in `raw_data`
- This data is still unstructured and needs processing

```python
#2-1 : temporary testing step: what is the structure of the API data?
city = input("Enter city: ")
api_key = input("Enter API key: ")

raw_data = get_weather(city, api_key)

import json
print(json.dumps(raw_data, indent=4))
```

## Step 3: Extract Only Required Data

### What this step does

The API returns a large JSON response with a lot of information.

In this step, we extract only the important fields:
- city name
- temperature
- weather condition
- humidity

This makes the data easier to use and display.

### How it works

- A function `parse_weather_data()` is created
- It takes raw API data as input
- Specific values are extracted from nested JSON fields:
  - `data["name"]` → city name
  - `data["main"]["temp"]` → temperature
  - `data["weather"][0]["description"]` → condition
  - `data["main"]["humidity"]` → humidity
- A new dictionary is returned with only the required information

This step converts raw API data into a clean and usable format.

```python
#3 : parse only what you need
def parse_weather_data(data):
    return {
        "city": data["name"],
        "temperature": data["main"]["temp"],
        "condition": data["weather"][0]["description"],
        "humidity": data["main"]["humidity"]
    }
```

## Step 4: Display Weather Information

### What this step does

This step takes the processed weather data and displays it in a clean, readable format.

Instead of showing raw JSON, the output is formatted for easy understanding.

### How it works

- A function `display_weather()` is created
- It takes the cleaned weather data as input (`summary`)
- Each value is formatted and printed:
  - city name
  - temperature
  - weather condition (formatted using `.title()`)
  - humidity

This ensures the final output is user-friendly and easy to read.

```python
#3: Function to format output
def display_weather(summary):
    print(f"Weather in {summary['city']}:")
    print(f"Temperature: {summary['temperature']}°C")
    print(f"Condition: {summary['condition'].title()}")
    print(f"Humidity: {summary['humidity']}%")
```

## Step 5: Run the Complete Application

### What this step does

This step connects all the functions together and runs the full weather application.

It takes user input, fetches data from the API, processes it, and displays the result.

### How it works

- The user enters a city name and API key
- `get_weather()` fetches raw data from the API
- If data is successfully received:
  - `parse_weather_data()` extracts the required information
  - `display_weather()` shows the final output

This creates a complete data flow from input → API → processing → output.

```python
#5: input & Run flow
city = input("Enter city: ").strip()
api_key = input("Enter your API key: ").strip()

raw_data = get_weather(city, api_key)

if raw_data:
    weather_summary = parse_weather_data(raw_data)
    display_weather(weather_summary)
```
