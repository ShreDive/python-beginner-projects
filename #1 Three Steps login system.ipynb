# Python Login System

This notebook demonstrates a basic Python login system built in three stages:

1. Single-attempt login  
2. Login system with 3 retry attempts  
3. Login system with attempt countdown and account lock  

The purpose of this project is to practice:
- variables
- user input
- conditionals
- while loops
- basic authentication logic

## Project Overview

This mini project shows how a simple login system can be improved step by step.

Each version adds a new feature:
- Version 1 checks credentials once
- Version 2 allows up to 3 attempts
- Version 3 shows how many attempts are left before lockout

## Step 1: Single-Attempt Login

In this version, the user gets only one chance to enter the correct username and password.
If either value is incorrect, login fails immediately.

```python
# STEP 1: Single-attempt login

correct_username = "user@example.com"
correct_password = "secure123"

input_username = input("Username: ")
input_password = input("Password: ")

if correct_username == input_username and correct_password == input_password:
    print("Login successful")
else:
    print("Invalid credentials")
```

### Logic Used
- Stored the correct username and password in variables
- Took user input using `input()`
- Used an `if-else` statement to compare the entered values
- Displayed success or failure message

## Step 2: Add Retry Attempts

This version improves the login system by allowing the user up to 3 attempts.
If the user enters the correct credentials before reaching the limit, login succeeds.
Otherwise, the account is locked.

```python
# STEP 2: Add retry attempts (3 tries)

correct_username = "user@example.com"
correct_password = "secure123"

set_attempts = 0

while set_attempts < 3:
    input_username = input("Username: ")
    input_password = input("Password: ")

    if input_username == correct_username and input_password == correct_password:
        print("Login successful")
        break
    else:
        print("Incorrect. Try again.")
        set_attempts += 1

if set_attempts == 3:
    print("Account locked.")
```

### Logic Used
- Initialized an attempts counter
- Used a `while` loop to allow repeated login attempts
- Stopped the loop using `break` when login succeeded
- Locked the account after 3 failed attempts

## Step 3: Show Attempts Left

This final version makes the login system more user-friendly by displaying how many attempts remain after each failed login.

```python
# STEP 3: Show how many attempts are left

correct_username = "user@example.com"
correct_password = "secure123"

set_attempts = 0

while set_attempts < 3:
    input_username = input("Username: ")
    input_password = input("Password: ")

    if input_username == correct_username and input_password == correct_password:
        print("Login successful")
        break
    else:
        set_attempts += 1
        print(f"Incorrect. Try again. Attempts left: {3 - set_attempts}")

if set_attempts == 3:
    print("Account locked.")
```
